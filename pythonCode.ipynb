```python
#!/usr/bin/python3
# Author: Prasandeep Singh

# Import packages/modules
import tkinter as tk
from tkinter.constants import *
from tkinter import messagebox 

# Creation of our base class

class myProjectClass():
    outputList =  []    # datatype "list" globally decalerd for the output  
            
    def __init__(self, master):     # Initialization of application; attributes and widgets
        master.title('Python Project')
        master.geometry("780x350")
        
        
        self.myProjectLabel1 = tk.Label(master, text = "Enter File Name:") #creation of label widget 
        self.myProjectLabel1.grid(row=0, column=0,padx=5, pady=5)  #placing the widget on Window
        
        self.myEntrySpace = tk.Entry(master, text="Entry wideget data") #creation of entry widget
        self.myEntrySpace.grid(row=0, column=1,padx=5, pady=5)  #placing the entry widget on window
        
        self.myFetchFileButton = tk.Button(master, text="Fetch File", command= self.myFetchFileClick) #button to fetch file
        self.myFetchFileButton.grid(row=0, column=2,padx=5, pady=5) #placing button on window
        
        self.myProcessFileFileButton = tk.Button(master, 
                                       text="Process selected file", 
                                       command= self.myProcessFileClick) #button widget to process the file
        self.myProcessFileFileButton.grid(row=1, column=0,padx=5, pady=5) #placing the button
        
        self.myProjectLabel2 = tk.Label(master, text = "") #label to show status if file run successfully
        self.myProjectLabel2.grid(row=1, column=1,padx=5, pady=5) #placing the label on window
        
        self.myProjectLabel3 = tk.Label(master, text = "") #label to show output path on our window
        self.myProjectLabel3.grid(row=1, column=2,padx=5, pady=5) #placing the label on window
        
        self.myTextSpace = tk.Text(master, width = 40, height = 5, wrap = WORD) #text widget to show output on window
        self.myTextSpace.grid(row=2, column=1,padx=5, pady=5) #placing the text widget on window
        
        self.myWindowExitButton = tk.Button(master, text="Exit",command= root.destroy) #adding exit button widget
        self.myWindowExitButton.grid(row=5, column=1,padx=5, pady=5) #placing the widget on window
        


    # Creation of Methods 
        
    def myProcessFileClick(self):   
        self.myWordsCountAndCharactersCounts() #call this method
        self.mySpaceCountFunct()    #call this method
        self.myWordOccurenceCountFunct()  #call this method
        
        self.myProjectLabel2.config(text="Successfully proccessed the given file")  #configuring the text to show that file has been processed successfully
        
        with open(self.outputFilePath, 'r' ):  #opens our text file in read mode
            self.myTextSpace.insert(1.0,self.outputList)   #and insert contents of output file on our text widget
        outputFilePath = "Output file location:" + "C:\\MyProjectFiles\\myOutput" # show the location 
        self.myProjectLabel3.config(text=outputFilePath)                         # of output file on widget
        
    
    def myWordsCountAndCharactersCounts(self):
        wordsCount = 0
        charactersCount = 0
        varOne = self.myFileText  #varOne is variable that has content of our file
        for line in varOne:     #for loop to count words and characters
            wordsCountlist = line.split()
            wordsCount = wordsCount + len(wordsCountlist)
            charactersCount = charactersCount + len(line)
            
        self.outputList.append("You are viewing the output analysis of:")
        self.outputList.append(self.myEntrySpace.get())
        self.outputList.append("\n")
        
        self.outputList.append("Number of words:") # text for output appended in our output file
        self.outputList.append(wordsCount) #output is being appended on our output file
        self.outputList.append(" ")
        self.outputList.append("Number of characters:")
        self.outputList.append(charactersCount)
        self.outputList.append("\n")
        print("\n")
        print("Number of words: ", wordsCount)
        print("number of characters: ",charactersCount)
        self.CharactersCounts = charactersCount #variable is created to calculate the percentage of blank spaces in different method
        
        
        
    def mySpaceCountFunct(self): 
        mySpaceCount = 0
        varTwo = self.myFileText 
        for line in varTwo:   #loop to find blank spaces
            for letter in line:
                if letter.isspace(): #isspace is used to find the whitespace character
                    mySpaceCount=mySpaceCount+1
        print("Number of spaces: ", mySpaceCount)
        self.spaces  = mySpaceCount
        
        try:
            self.percentage = round((self.spaces *100)/self.CharactersCounts) #percentage of blank spaces is calculated over total characters
            self.outputList.append("Blank space characters:")
            self.outputList.append(self.spaces)
            self.outputList.append("\n")
            self.outputList.append("Percentage of blank spaces: ")
            myPercentage = str(self.percentage) + "%"
            self.outputList.append(myPercentage)
            self.outputList.append("\n")
            nonBlankCharacters = self.CharactersCounts - self.spaces
            print("Non blank characters are:", nonBlankCharacters )
            self.outputList.append("Non blank characters are:")
            self.outputList.append(nonBlankCharacters)
            self.outputList.append("\n")
            print("Percentage of blank spaces: ", self.percentage, "%" )
            print("\n")

        except:
            print("Please check your input")
            messagebox.showinfo("Warning!!","Please check your input")
        
        
        
    def myWordOccurenceCountFunct(self):
        wCounter = {}  #initialization of dictionary to count words occurance 
        varThree = self.myFileText

        for line in varThree:

            punctuation_list = line.replace(',','').replace('\'','').replace('.','').replace(':','').lower().split()
            punctuation_list = line.replace(';','').replace('/','').replace('----','').replace('--','').lower().split() #replace the punctuation with spaces
            punctuation_list = line.replace('<','').replace('>','').replace('^','').replace('-','').lower().split()
            for word in punctuation_list:
      
                if word not in wCounter:
                    wCounter[word] = 1
                else:
                    wCounter[word] = wCounter[word] + 1

        print('{:60}{:75}'.format('Word','Frequency')) #format of printing down
        print('*' * 70)
        self.outputList.append("The number of words and their counts")
        self.outputList.append("\n")


        for  (word,frequency)  in wCounter.items(): 
            print('{:50}{:18}'.format(word,frequency))
            self.outputList.append(word)
            self.outputList.append("Its frequency:")
            self.outputList.append(frequency)
            self.outputList.append("\n")
        self.outputFilePath = "c:\\\\MyProjectFiles\\\\myOutput.txt" #path where output file is stored
        with open(self.outputFilePath, 'w+' ) as file:
            for item in self.outputList:
                file.write("%s\n" %item)
                
                
        
        

    def myFetchFileClick(self):   #load the file
        print(self.myEntrySpace.get()) #shows name of the file being read
        self.nameOfFile = self.myEntrySpace.get()
        self.myFilePath = "c:\\\\MyProjectFiles\\\\" + self.nameOfFile  #stores the path of file in a variable 
        print("File path: ",self.myFilePath) #shows the file path
        try:
            self.textFilePath = open(self.myFilePath)
        except FileNotFoundError:
            print("No Such file Exist!")
            return
        self.myFileText = self.textFilePath.readlines() # read the our file texts and store that in a variable


# Creating application
root = tk.Tk()   
app = myProjectClass(root)

# Starting application
root.mainloop()
```
